```python
# --- ASSISTENTE PEDAGÓGICO COM IA ---

import os
from google.colab import userdata
from google import genai

print("Configurando o ambiente e inicializando a IA...")
os.environ['GOOGLE_API_KEY'] = userdata.get('alura') # Verifique se 'alura' é o nome do seu secret
client = genai.Client()
print("Ambiente pronto!")

print("Criando a base de dados dos alunos...")
lista_de_nomes = ["Maria Silva", "João Santos", "Ana Oliveira", "Pedro Costa", "Juliana Pereira"]
lista_de_medias = [8.9, 7.5, 4.2, 1.4, 9.5]
medias_atuais = lista_de_medias.copy()
print("Base de dados pronta!")

while True:
    print("\n--- MENU: Assistente Pedagógico com IA ---")
    print("1. Ver Boletim da Turma")
    print("2. Aplicar Ponto de Bônus")
    print("3. Gerar Plano de Estudos com IA")
    print("4. Sair")

    escolha = input("Escolha uma opção: ")

    if escolha == '1':
        print("\n--- Boletim da Turma ---")
        boletim_completo = []
        for i in range(len(lista_de_nomes)):
            aluno_info = {"nome": lista_de_nomes[i], "media": medias_atuais[i]}
            boletim_completo.append(aluno_info)
        for aluno in boletim_completo:
            print(f"Aluno(a): {aluno['nome']: <20} | Média Atual: {aluno['media']:.1f}")

    elif escolha == '2':
        print("\nAplicando 1.0 ponto de bônus em todas as médias...")
        n = 0
        while n < len(medias_atuais):
            medias_atuais[n] = medias_atuais[n] + 1.0
            if medias_atuais[n] > 10.0:
                medias_atuais[n] = 10.0
            n = n + 1
        print("✅ Bônus aplicado com sucesso! Consulte o boletim para ver as novas notas.")

    elif escolha == '3':
        print("\n--- Gerador de Plano de Estudos com IA ---")
        for i in range(len(lista_de_nomes)):
            print(f"{i}: {lista_de_nomes[i]} | Média: {medias_atuais[i]:.1f}")

        try:
            indice_aluno = int(input("\nDigite o NÚMERO do aluno para gerar o plano: "))
            if 0 <= indice_aluno < len(lista_de_nomes):
                aluno_nome = lista_de_nomes[indice_aluno]
                aluno_media = medias_atuais[indice_aluno]
                materia = "Matemática"

                print(f"\nIniciando assistente para o(a) aluno(a): {aluno_nome}, Média: {aluno_media}")
                chat = client.chats.create(model="gemini-2.5-flash")

                prompt_inicial = f"""
                Aja como um(a) especialista em pedagogia. O(A) aluno(a) '{aluno_nome}' está com a média {aluno_media} em {materia}.
                Crie um plano de estudos de recuperação simples e encorajador com 2 atividades práticas e 1 dica de estudo.
                """
                resposta = chat.send_message(prompt_inicial)
                print(f"\n--- Plano Sugerido pela IA ---\n{resposta.text}\n")

                print("--- Chat aberto para refinar o plano ---")
                prompt_professor = input("Digite suas perguntas ou ajustes (ou 'fim' para encerrar): ")
                while prompt_professor.lower() != "fim":
                    resposta = chat.send_message(prompt_professor)
                    print(f"\nIA: {resposta.text}\n")
                    prompt_professor = input("Digite suas perguntas ou ajustes (ou 'fim' para encerrar): ")
            else:
                print("\n❌ Erro: Número fora do intervalo. Por favor, escolha um número da lista.")
        except ValueError:
            print("\n❌ Erro: Entrada inválida. Por favor, digite apenas o número do aluno.")

    elif escolha == '4':
        print("\nEncerrando o assistente. Bom trabalho, professor(a)!")
        break

    else:
        print("\n❌ Opção inválida! Por favor, escolha um número de 1 a 4.")
```
